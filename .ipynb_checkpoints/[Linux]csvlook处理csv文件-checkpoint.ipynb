## Csvcut

主要用于查看某些行列

调出第二列前十行

```python
!csvcut -c 2 test.csv | head -10
```

展示所有列名称

```python
!csvcut -n test.csv
```

```python
!csvcut -c 2,3,4 test.csv | head -10 | csvlook
```

```python
!csvcut -c ip,app,device test.csv| head -10|csvlook
```
